```python
import dlt
import json
```

```python
filesystem = "./dbfiles/mercadolivre"
```

```python
pipeline = dlt.pipeline(
    pipeline_name="ny_taxi_data",
    destination="duckdb",
    dataset_name="mercadolivre",
)
```

```python
data_path = "../data/raw/data.json"
with open(data_path, "r") as json_file:
    data = [json.load(json_file)]
```

```python
info = pipeline.run(data, table_name="mercadolivre", write_disposition="replace")
```

```python
print(info)
```
